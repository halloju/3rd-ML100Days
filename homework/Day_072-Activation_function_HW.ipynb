# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''

def relu(x):
    return np.maximum(0, x)

def drelu(x):
    return (x>=0)*1

X = np.linspace(-10,10,100)

plt.plot(X, relu(X), 'b', label='linspace(-10,10,10)')
plt.title('ReLU function')
plt.grid()
plt.text(4,0.8, '$f(x)=max(0,x)$')

plt.show()
```

```python
plt.plot(X, drelu(X), 'r')
plt.title('Derivitive ReLU function')
plt.grid()
plt.text(4,0.8, '$f(x)= I(x>=0)')

plt.show()
```
