- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
print(X.shape)
print(y.shape)
```

```python
unique, counts = np.unique(y, return_counts=True)
dict(zip(unique, counts))
```

```python
estimators = [('k_means_3', KMeans(n_clusters=3, n_init=10, init='random')),
              ('k_means_3', KMeans(n_clusters=3, n_init=20, init ='k-means++')),
              ('k_means_5', KMeans(n_clusters=5, n_init=20, init ='k-means++')),
              ('k_means_7', KMeans(n_clusters=7, n_init=20, init ='k-means++'))]
```

```python
fignum = 1
titles = ['3 clusters, random', '3 clusters, k++', '5 cluster, k++', '7 cluster, k++']
for name, est in estimators:
    f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
       
    ## fit data
    est.fit(X)
    
    labels = est.labels_

    ax1.scatter(X[:, 0], X[:, 1],
               c=labels.astype(np.float), edgecolor='k')
    ax2.scatter(X[:, 2], X[:, 3],
               c=labels.astype(np.float), edgecolor='k')
    ax1.set_title(titles[fignum-1])
    fignum = fignum + 1


fig = plt.figure(fignum, figsize=(4, 3))
f, (ax3, ax4) = plt.subplots(1, 2, sharey=True)

ax3.scatter(X[:, 0], X[:, 1], c = y)
ax4.scatter(X[:, 2], X[:, 3], c = y)
ax3.set_title('True')

```
